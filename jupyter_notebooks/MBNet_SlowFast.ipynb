**Unpack and prepare Jester dataset**

```
!mkdir -p "/content/datasets/jester"
%cd "/content/drive/My Drive/datasets/jester"
```

```
%cd "/content/drive/My Drive/datasets/jester"
import time
import os
from zipfile import ZipFile

UNPACK_DEST = "/content/datasets/jester"
VOLUME_DIR = "/content/drive/My Drive/datasets/jester"
for volume_ind in range(0, len(os.listdir(VOLUME_DIR))):
# for volume_ind in range(4, 5):
  time_start = time.time()

  new_vol_dir = os.path.join(UNPACK_DEST, "vol_{}".format(volume_ind))
  if not os.path.exists(new_vol_dir):
    os.mkdir(new_vol_dir)

  zip_volume_name = "vol_{}.zip".format(volume_ind)
  zip_volume_path = os.path.join(VOLUME_DIR, zip_volume_name)

  zip_obj = ZipFile(zip_volume_path, 'r')

  zip_obj.extractall(new_vol_dir)
  zip_obj.close()

  print("Completed vol {} extraction in {} seconds".format(volume_ind, time.time() - time_start))
```
